# PREDICT HOUSE PRICES

#### - Dataset includes house sale prices for king country in washington, USA.
#### - Homes that are sold in the time period: May, 2014, and May, 2015.
#### - Model Inputs
	- ida: notation for a house
	- date: Date house was sold
	- bedrooms: Number of Bedrooms
	- bathrooms: Number of bathrooms
	- sqft_living: home square footage
	- sqft_lot: square footage of the lot
	- floors: Total floors (levels) in house
	- waterfront: waterfront property
	- condition: How good the condition is (overall)
	- grade: overall grade given to housing unit, based on King Country grading System
	- sqft_abovesquare: Footage of house apart from basement
	- sqft_basement: square footage of the basement
	- yr_built: Built year
	- yr_renovated: Year when house was renovated
	- zipcode: zip
	- lat: Latitude coordinate
	- long: Longitude coordinate
	- sqft_living15: Living room area in 2015
	- sqft_lot15: lot size area in 2015 (implies-- some renovations)

#### The model should predict:
	- House Price

# STEP #3: CREATE TESTING AND TRAINING DATASET/DATA CLEANING

```python
X = house_df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement']]
```

```python
y = house_df['price']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
X_scaled.shape
```

```python
scaler.data_max_
```

```python
y = y.values.reshape(-1, 1)
```

```python
y_scaled = scaler.fit_transform(y)
```

```python
y_scaled
```

# STEP #4: TRAINING THE MODEL

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size = 0.25 )
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(units=100, activation='relu', input_shape=(7,) ))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=1, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss='mean_squared_error')
```

```python
epochs = model.fit(X_train, y_train, epochs = 100, batch_size = 50, validation_split=0.2 )
```

# STEP #5: EVALUATE THE MODEL

```python
epochs.history.keys()
```

```python
plt.plot(epochs.history['loss'])
plt.plot(epochs.history['val_loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epochs')
plt.ylabel('Training and Validation Loss')
plt.legend(['Training Loss','Validation Loss'])
```

```python
y_predict = model.predict(X_test)
```

```python
plt.plot(y_test, y_predict, '^', color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```

```python
#
X_test_1 = np.array([[4, 3, 1960, 5000, 1, 2000, 3000]])

scaler_1 = MinMaxScaler()
X_test_1_scaled = scaler_1.fit_transform(X_test_1)

y_predict_1 = model.predict(X_test_1_scaled)

y_predict_1 = scaler.inverse_transform(y_predict_1)
y_predict_1
```

```python
y_predict_origin = scaler.inverse_transform(y_predict)
y_test_origin = scaler.inverse_transform(y_test)
```

```python
plt.plot(y_test_origin, y_predict_origin, 'o', color = 'r')
plt.xlabel('Model Predictions')
plt.xlabel('True Values')
plt.xlim(0, 5000000)
plt.ylim(0, 3000000)
```

```python
K = X_test.shape[1]
n = len(X_test)
```

```python
K
```

```python
n
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt
```

```python
RMSE = float(format(np.sqrt(mean_squared_error(y_test_origin, y_predict_origin)), '.3f'))
MSE = mean_squared_error(y_test_origin, y_predict_origin)
MAE = mean_absolute_error(y_test_origin, y_predict_origin)
r2 = r2_score(y_test_origin, y_predict_origin)
adj_r2 = 1-(1-r2) * (n-1) / (n-K-1)
```

```python
print('RMSE = ', RMSE)
print('MSE = ', MSE)
print('MAE = ', MAE)
print('R2 = ', r2)
print('Adjusted R2 = ', adj_r2)
```
